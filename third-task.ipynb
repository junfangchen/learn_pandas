## 实操练习

1、 speed > 100 的pokemon 共有多少个？

2、90< speed < 100 的 pokemon 的 name 是哪些？


```python
# 答案
#print(df["Speed"])
alist = list(df["Speed"])
sum = 0
for i in alist:
    if i > 100:    # 对吗？为什么？自己试试代码，查看为什么。
        sum += 1
print(sum)
        
```

```python
len(df[(df['Speed']>100)]) #现在对吗？
```

```python
# 答案,符合&表示“且” ，符号|表示“或”，符号 ~表示“非”，注意不能用python原生的 （and or not） 来表示逻辑关系
dfx = df[(df['Speed']>90) & (df['Speed']<100)]


for i in dfx['Name']:
    print(i,end=', ')

print(len(dfx))
dfy =  df[(df['Speed']<=90) | (df['Speed']>=100)]
print(len(dfy))

df.loc[:,["Speed"]]
```
